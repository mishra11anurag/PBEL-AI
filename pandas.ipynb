```python
# input for city,team
def get_team_win_counts(dataframe, team_name, specific_city=None):

    total_wins_all_cities = dataframe[dataframe['winner'] == team_name].shape[0]


    wins_in_specific_city = 0
    if specific_city:

        wins_in_specific_city = dataframe[
            (dataframe['winner'] == team_name) &
            (dataframe['city'].str.lower() == specific_city.lower())
        ].shape[0]

    return wins_in_specific_city, total_wins_all_cities


team_example = 'Chennai Super Kings'
city_example = 'Mumbai'



wins_in_city_count, total_wins_count = get_team_win_counts(df, team_example, city_example)


if total_wins_count is not None:
    print(f"\nResults for '{team_example}':")
    print(f"Total times '{team_example}' won in '{city_example}': {wins_in_city_count} times")
    print(f"Total times '{team_example}' won across all cities: {total_wins_count} times")
else:
    print(f"Could not calculate results for '{team_example}'. Please check team name or data.")


print("\n" + "="*50 + "\n")


print("probability",wins_in_city_count/total_wins_count)
```
